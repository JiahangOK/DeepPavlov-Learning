# 第一个HelloBot程序

这个 HelloBot 程序会创建一个简单的基于模式匹配的对话机器人  
## 完整代码：
~~~python
from deeppavlov.skills.pattern_matching_skill import PatternMatchingSkill
from deeppavlov.core.agent import Agent, HighestConfidenceSelector

hello = PatternMatchingSkill(responses=['Hello world! :)'], patterns=["hi", "hello", "good day"])
bye = PatternMatchingSkill(['Goodbye world! :(', 'See you around.'], ["bye", "chao", "see you"])
fallback = PatternMatchingSkill(["I don't understand, sorry :/", 'I can say "Hello world!" 8)'])

HelloBot = Agent([hello, bye, fallback], skills_selector=HighestConfidenceSelector())

HelloBot(['Hello!', 'Boo...', 'Bye.'])

~~~


## 解读：  
导入deeppavlov库

~~~python
from deeppavlov.skills.pattern_matching_skill import PatternMatchingSkill
from deeppavlov.core.agent import Agent, HighestConfidenceSelector
~~~

模式匹配是自然语言理解组件中最简单的例子,通过文本找到已经定义的模式  
我们先定义三个简单的模式：  
hello:当问到关于"hi","hello","good day"相关的话时，对话机器人会自动回复"Hello world! :)",其中不分大小写  
  机器人的patterns是人和它说的，responce是机器人回复的内容。  
  如果PatternMatchingSkill只有一个链表参数，那么这属于机器人无法匹配模式的内容，是机器人的默认回复。
bye和fallback同理。  
hello,bye,fallback是对话机器人的三个skill。
~~~python
hello = PatternMatchingSkill(responses=['Hello world! :)'], patterns=["hi", "hello", "good day"])
bye = PatternMatchingSkill(['Goodbye world! :(', 'See you around.'], ["bye", "chao", "see you"])
fallback = PatternMatchingSkill(["I don't understand, sorry :/", 'I can say "Hello world!" 8)'])
~~~

当在对机器人说的话里找到PatternMatchingSkill里的patterns含有的内容时，返回的confidence是1；  
如果没找到，那么返回的confidence是0.5。  

Agent是对话的管理员，需要向Agent提供所有skill和skills_selector  
skill也就是之前的那些模式，是链表的形式  
skills_selector中最简单的就是HighestConfidenceSelector(),它会选择最大的confidence的skill的回复作为回复

~~~python
HelloBot = Agent([hello, bye, fallback], skills_selector=HighestConfidenceSelector())
~~~


这个对话过程在这个DeepPavlov库中是批量工作的，我们可以向机器人传递一批请求，如下：
~~~python
HelloBot(['Hello!', 'Boo...', 'Bye.'])
~~~

输出结果：
~~~
['Hello world! :)', "I don't understand, sorry :/", 'See you around.']
~~~
